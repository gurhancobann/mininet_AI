```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("dataTest23.06.2024.csv")
df=pd.DataFrame(df)
df.head()
```

```python
df["mos"]=6
df.shape[0]
for i in range(0,df.shape[0],1):
    if df.ssim_first[i]>=0.99:
        df.mos[i]=5
    elif df.ssim_first[i]>=0.95 and df.ssim_first[i]<0.99:
        df.mos[i]=4
    elif df.ssim_first[i]>=0.88 and df.ssim_first[i]<0.95:
        df.mos[i]=3
    elif df.ssim_first[i]>=0.5 and df.ssim_first[i]<0.88:
        df.mos[i]=2
    elif df.ssim_first[i]<0.5:
        df.mos[i]=1

df
```

```python
df.dropna(inplace=True)
df.describe().T
```

```python
y=df["mos"]
x=df.drop(["mos","host","ssim_first","ssim_second","type","psnr","server","packetLoss"],axis="columns")
y
```

```python
x
```

```python
XTrain,XTest,yTrain,yTest=train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
```

```python
model.add(Dense(20,input_dim=7,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['r2_score'])
```

```python
model.fit(XTrain,yTrain,batch_size=30,epochs=200)
```

```python
test_results=model.evaluate(XTest,yTest,batch_size=30)
```

```python
print("test loss, test acc:",test_results)
```
